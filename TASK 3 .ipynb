# Import Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('whitegrid')
```

# Read the dataset

```python
df = pd.read_csv('SampleSuperstore.csv', low_memory = False)
df.head()
```

```python
# shape of the dataset
df.shape
```

In this given dataset we have 9994 rows and 13 columns.

```python
# checking information of the dataset
df.info()
```

```python
# checking null value 
df.isnull().sum()
```

There are no any missing values in any of the column.

```python
df.describe() # getting only for numerical variable
```

 The output results provide the five number summary of the data.

```python
# Find unique value of all the variables
df.nunique()
```

```python
#Check the measure of dispersion

df.quantile(0.75) - df.quantile(0.25)
```

```python
# the above output represents the IQR values for all the attributes.
```

# Heatmap for checking the correlation

```python
# prints the correlation coefficient between every pair of attributes
df.corr()
```

```python
fig,ax = plt.subplots(figsize=(10, 10))   
sns.heatmap(df.corr(), ax=ax, annot=True, linewidths=0.05, fmt= '.2f',cmap="magma")
plt.show()
```

From the above plot we analyse that sales and profit are more correlated to each other.

# Distribution of Region

```python
sns.countplot(df['Region'], order = df['Region'].value_counts().index)
```

# Distribution of State

```python
plt.figure(figsize=(20,9))
plt.xticks(rotation=90)
sns.countplot(df['State'], order=df['State'].value_counts().index)
```

# Distribution of Category

```python
sns.countplot(x=df['Category'] , order=df['Category'].value_counts().index)
```

# Distribution of Segment

```python
sns.countplot(x=df['Segment'], order=df['Segment'].value_counts().index)
```

# Distribution of Subcategory

```python
plt.figure(figsize=(12,5))
plt.xticks(rotation=90)
sns.countplot(x=df['Sub-Category'], order=df['Sub-Category'].value_counts().index)
```

# Visualize the two categorical variables and one numerical variable

```python
plt.figure(figsize = (15,5))
sns.barplot(data=df, x = 'Region', y='Sales', hue ='Segment')
plt.show()
```

From the above plot we can analyse that the number of sales of home office in south is higher than other segment in all
the  other regions.

```python
plt.figure(figsize =(15, 5))
sns.barplot(data=df, x='Region', y='Profit', hue='Segment')
plt.show()
```

The above plot shows that profit in consumer segment is ver less in cental region.

```python
plt.figure(figsize=(15,5))
sns.barplot(data=df, x= 'Region', y='Profit', hue='Category')
plt.show()
```

From the above plot we can analyse that profit on furniture in central region is very less amount.
and profit on technology in east region is higher than all the region .

```python
plt.figure(figsize=(15,5))
sns.barplot(data=df, x='Region', y='Sales', hue='Category')
plt.show()
```

The above plot shows that  in south region the sales of technology category is higher than  all  the other regions. 

# Statewise sales and profit distribution

```python
Result = df.groupby('State')['Sales', 'Profit'].sum()
Result.plot(kind='bar', figsize=(25,7))
```

Hence the above plot shows that  the sales is high in California  state than the other sates.
